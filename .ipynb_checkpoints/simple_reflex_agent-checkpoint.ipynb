# Phase 2: Sense Logic #
<p4>
    <ul>
        <li>Create a single function that abstracts perception</li>
        <li>Given an image path → returns the predicted class and confidence score.</li>
        <li>This becomes your agent’s eyes — nothing more.</li>
    </ul>
</p4>

```python
#------------Loading the Model----------------------
def loading_model(model_Name):
    model_Name = load_model('full_model.keras')
    return model_Name

#---- Calling the Function ----------------------
model = loading_model('full_model.keras')    
```

```python
BASE_DIR = Path('split/val')
image_paths = []
for classes in os.listdir(BASE_DIR):
    class_path = os.path.join(BASE_DIR, classes)
    for filename in os.listdir(class_path):
        image_path = os.path.join(class_path, filename)
        image_paths.append(image_path)

len(image_paths)
```

```python
#----------Writing the Sense Logic------------------------
#--- This function is for  for demo, conceptual clarity, and visuals ------------------
def percept_logique(image_path, model_Name):
    #---- Starting the preprocessu=ing of the image--------------
    image = load_img(image_path, target_size=(224,224))
    image_array = img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)
    preprocess_image = preprocess_input(image_array)

    # Predicting
    prediction = model_Name.predict(preprocess_image)
    prediction_index = np.argmax(prediction[0])
    class_names = ['Cataract', 'Diabetic_retinopathy', 'Glaucoma', 'Normal']
    predicted_class = class_names[prediction_index]

    plt.imshow(image)
    plt.axis('off')

    plt.title(f"Predicted: {predicted_class}")
    plt.show()

    confidence = prediction[0][prediction_index]

    print(f"The image is classified as: {predicted_class}")
    print(f'The Confidence is: {confidence}')

    return predicted_class, confidence

percept_logique(image_path=image_paths[0], model_Name=model)
```

```python
#-----------The sense() → This is the modular logic to be used in agent_loop.py and sensor.py----------------------------

def sense(image_path, model_Name):
    #---- Starting the preprocessu=ing of the image--------------
    image = load_img(image_path, target_size=(224,224))
    image_array = img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)
    preprocess_image = preprocess_input(image_array)

    # Predicting
    prediction = model_Name.predict(preprocess_image)
    prediction_index = np.argmax(prediction[0])
    class_names = ['Cataract', 'Diabetic_retinopathy', 'Glaucoma', 'Normal']
    predicted_class = class_names[prediction_index]

    confidence = prediction[0][prediction_index]

    return predicted_class, confidence

sense(image_path=image_paths[0], model_Name=model)
```

# Phase 3: Decision Logic #
<p4> 
    <ul>
        <li>This phase is where agent decides what to do based on the sensory input.</li>
        <li>This is not just about making predictions — It is about building logic that chooses an action (e.g., Refer, Monitor, Ignore, etc.) based on the prediction and its confidence.</li>

            <li>Evolution Point---> This rule-based logic can later evolve into a utility-based agent or learning agent (e.g., Q-learning).</li>
        </ul>
    </ul>
</p4>

<h3>Actionable Decision Matrix for Fundus Disease Classification Using Model Confidence Thresholds</h3>

| Class                    | Clinical Urgency | High Confidence (≥ 0.85)     | Medium Confidence (0.4–0.85)         | Low Confidence (< 0.4)  |
| ------------------------ | ---------------- | ---------------------------- | ------------------------------------ | ----------------------- |
| **Glaucoma**             | High             | Refer immediately            | Refer¹                               | Monitor / Refer¹        |
| **Diabetic Retinopathy** | Moderate         | Refer for further evaluation | Monitor closely; schedule follow-up² | Monitor                 |
| **Cataract**             | Low              | Schedule surgical consult³   | Monitor; schedule routine follow-up  | Ignore / Routine screen |
| **Normal**               | None             | Ignore                       | Ignore                               | Ignore                  |


```python
def decision(predicted_class, confidence):
    if predicted_class == 'Glaucoma':
        if confidence >= 0.85:
            return 'Refer immediately'
        elif confidence >= 0.4:
            return 'Refer'
        else:
            return 'Monitor / Refer'

    elif predicted_class == 'Diabetic_retinopathy':
        if confidence >= 0.85:
            return 'Refer for further evaluation'
        elif confidence >= 0.4:
            return 'Monitor closely; schedule follow-up'
        else:
            return 'Monitor'

    elif predicted_class == 'Cataract':
        if confidence >= 0.85:
            return 'Schedule surgical consult'
        elif confidence >= 0.4:
            return 'Monitor; schedule routine follow-up'
        else:
            return 'Ignore / Routine screen'

    elif predicted_class == 'Normal':
        return 'Ignore'

    else:
        return 'Monitor'  # fallback for any unknown class
```

```python
decision(predicted_class, confidence)
```

# Phase 4: Implementing the Agent Loop #
<p4>

This is where you glue together your sensor and decision modules (and later, your environment simulator) into a coherent agent that:

1. Perceives the world
2. Decides what to do
3. Acts (logs or simulates an outcome)
4. Loops over a batch of inputs

</p4>

```python
def agent_loop(image_paths, model):
    logs = []  # ← Logging starts here

    for image_path in image_paths:
        predicted_class, confidence = sense(image_path, model)
        action = decision(predicted_class, confidence)  

        logs.append({
            "image_path": image_path,
            "predicted_class": predicted_class,
            "confidence": confidence,
            "action": action
        })

    return logs
```

```python
logs = agent_loop(image_paths, model)
```

```python
df = pd.DataFrame(logs)
```

```python
df
```

```python
df['true_label'] = df['image_path'].apply(lambda x: x.split('\\')[2].capitalize())
```

# Phase 6: Notebook Logging and Evaluation.#

<ul>
    <li>It contains Reward Logic: simulates a lightweight scoring function</li>
</ul>

```python
def get_reward(true_label, predicted_class, action):
    if true_label == predicted_class:
        if true_label == 'Normal' and action == 'Ignore':
            return 1  # Correctly ignored
        elif true_label != 'Normal' and 'Refer' in action or 'Consult' in action:
            return 1  # Referred appropriately
        elif 'Monitor' in action:
            return 0.5  # Somewhat cautious, acceptable
    return 0  # Mismatch

df['reward'] = df.apply(lambda row: get_reward(row['true_label'], row['predicted_class'], row['action']), axis=1)
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['reward'], bins=[0, 0.5, 1], kde=False)
plt.title("Reward Distribution")
plt.xlabel("Reward")
plt.ylabel("Number of Samples")
plt.xticks([0, 0.5, 1])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='true_label', hue='action')
plt.title("Actions Taken per True Label")
plt.xlabel("True Label")
plt.ylabel("Count")
plt.legend(title="Action", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
df['correct'] = df['true_label'] == df['predicted_class']
```
